```python
import matplotlib.pyplot as plt

from torch.optim.lr_scheduler import ExponentialLR, LinearLR
from torch import nn
from torch.optim import Adam
```

```python
optimizer = Adam(nn.Linear(1,1).parameters(), lr=1e-3)
scheduler = ExponentialLR(
    optimizer=optimizer,
    gamma = 1-1.75e-3
)

lr = [scheduler.get_last_lr()[-1]]
for _ in range(2000):
    scheduler.step()
    lr.append(scheduler.get_last_lr()[-1])


optimizer = Adam(nn.Linear(1,1).parameters(), lr=1e-3)
scheduler_linear = LinearLR(
    optimizer,
    start_factor=1.0,
    end_factor=0.02,
    total_iters=1500,
)
lr_linear = [scheduler_linear.get_last_lr()[-1]]
for _ in range(1500):
    scheduler_linear.step()
    lr_linear.append(scheduler_linear.get_last_lr()[-1])

print(lr[800])
print(lr[1000])
print(lr[-1])
fig = plt.figure()
plt.plot(lr)
plt.plot(lr_linear, label='linear')
plt.legend()
plt.show()

fig = plt.figure()
plt.plot(lr)
plt.plot(lr_linear, label='linear')
plt.yscale('log')
plt.legend()
plt.show()
```
